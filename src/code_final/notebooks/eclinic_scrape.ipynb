```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
driver = webdriver.Chrome()
driver.get("https://eclinic.mn/patient/search/hospital")
time.sleep(5)
```

```python
# Hospital webscrape code
data = []
total_pages = 18  
for page in range(1, total_pages+1 ):
    try:
        page_link = WebDriverWait(driver, 10).until( EC.element_to_be_clickable((By.XPATH, f"//li[@title='{page}']")))
        driver.execute_script("arguments[0].scrollIntoView();", page_link)
        page_link.click()
        if page<2:
            print(f"Clicked page {page}")
        time.sleep(2)  
        div_elements = driver.find_elements(By.CSS_SELECTOR, "div.tw-grid.tw-gap-2")
        for div in div_elements:

            hospital_name_element = div.find_element(By.CSS_SELECTOR, "div.tw-text-doctorText.tw-font-bold")
            hospital_name = div.find_element(By.CSS_SELECTOR, "div.tw-text-doctorText.tw-font-bold").text
            hospital_type = div.find_element(By.CSS_SELECTOR, "div.tw-text-patientSecondary.tw-font-medium").text
            address = div.find_element(By.CSS_SELECTOR, "div.tw-flex.tw-items-center.tw-pt-2.tw-text-patientText.tw-text-sm").text
            print(hospital_name)
            hospital_name_element.click()
            time.sleep(3)
        
            try:
                more_info_span = driver.find_element(By.XPATH, "//span[text()='Цааш харах']")
                print("Found 'Цааш харах' span. Clicking it.")
                more_info_span.click()
                time.sleep(3)
                activity_div = driver.find_element(By.CSS_SELECTOR, 'div.tw-p-4.tw-mb-4.tw-rounded-lg.tw-bg-white.tw-shadow')
                list_items = activity_div.find_elements(By.CSS_SELECTOR, 'ul.tw-list-disc li')
                activities = [item.text for item in list_items]      
                element = driver.find_element(By.CSS_SELECTOR, '.ant-dropdown-link')
                actions = ActionChains(driver)
                actions.move_to_element(element).click().perform()
                element.click()

            except:
                activity_div = driver.find_element(By.CSS_SELECTOR, 'div.tw-p-4.tw-mb-4.tw-rounded-lg.tw-bg-white.tw-shadow')
                list_items = activity_div.find_elements(By.CSS_SELECTOR, 'ul.tw-list-disc li')
                activities = [item.text for item in list_items]      
                element = driver.find_element(By.CSS_SELECTOR, '.ant-dropdown-link')
                actions = ActionChains(driver)
                actions.move_to_element(element).click().perform()
                element.click()

            data.append((hospital_name, hospital_type, address, activities))
    
    except Exception as e:
        df = pd.DataFrame(data, columns=["Hospital Name", "Hospital Type", "Address", "Activities"])
        df.to_csv('hospital_info.csv', index=False)
        break  
df = pd.DataFrame(data, columns=["Hospital Name", "Hospital Type", "Address", "Activities"])
df.to_csv('hospital_info.csv', index=False)
```
